```python
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np

%matplotlib inline
```

```python
MHW_intensity = np.linspace(0, 10)
performance_det = MHW_intensity * 1.3 + np.random.normal(size=len(MHW_intensity), scale=4) + 20
```

```python
performance_det
```

```python
sns.set(style="ticks")

plt.style.use('dark_background')
fig = plt.figure(figsize=(5,5), dpi=300)
ax = plt.axes()
sns.regplot(MHW_intensity, performance_det, fit_reg=True, color='white')
sns.despine()
plt.xlabel("MHW Intensity [Degree Days]")
plt.ylabel("Cumulative Performance Detriment")
plt.annotate("(Synthetic Data)", (0.65, 0.02), xycoords='axes fraction', alpha=0.4)
plt.xticks([], [])
plt.yticks([], [])
plt.savefig("synthetic_results.png", dpi=300, bbox_inches='tight')
```
